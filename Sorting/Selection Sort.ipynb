## Selection Sort 

[1,25,20,50,45,10,4] \
Step0: [1, 25, 20, 50, 45, 10, 4] \
Step1: [1, 4, 25, 50, 45, 20, 10] \
Step2: [1, 4, 10, 50, 45, 25, 20] \
Step3: [1, 4, 10, 20, 50, 45, 25] \
Step4: [1, 4, 10, 20, 25, 50, 45] \
Step5: [1, 4, 10, 20, 25, 45, 50] 

```python
li=[1,25,20,50,45,10,4]
n=len(li)
count=0
for i in range(n-1):
    for j in range(i,n):
        if li[i]>li[j]:
            li[i],li[j]=li[j],li[i]
    print("Step"+str(count)+": " + str(li))
    count+=1
```
