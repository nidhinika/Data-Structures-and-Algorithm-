# Quick Sorting Concepts

## Partition Algorithm 

```python
def partition_algo(arr):
    key=arr[-1]
    left=0
    for i in range(len(arr)):
        if arr[i]<=key:
            arr[left],arr[i]=arr[i],arr[left]
            left+=1
    print(arr)
    print("left",left)
    if left<len(arr):
        arr[left],arr[-1]=arr[-1],arr[left]
    return arr
partition_algo([2,5,1,0,4,6,3])
```

## Quick Sort Implemnation 

```python
def quick_Sort(arr):
    if len(arr)<=1:
        return arr
    l=partition_algo(arr)
    mid=len(l)//2
    left=quick_Sort(l[:mid])
    right=quick_Sort(l[mid:])
    return left+right
```

```python
def quick_sort(arr):
    # Base case: arrays with 0 or 1 element are already sorted
    if len(arr) <= 1:
        return arr
    
    # Choose pivot (here: last element)
    pivot = arr[-1]
    
    # Partition into three lists
    less = [x for x in arr[:-1] if x <= pivot]
    greater = [x for x in arr[:-1] if x > pivot]
    
    # Recursively sort and combine
    return quick_sort(less) + [pivot] + quick_sort(greater)

# Example usage
data = [10, 7, 8, 9, 1, 5]
print("Sorted array:", quick_sort(data))
```
