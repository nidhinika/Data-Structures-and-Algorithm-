## Insertion Sort 

li=[20,10,40,15,5] \
Step0: [10, 20, 40, 15, 5] \
Step1: [10, 20, 40, 15, 5] \
Step2: [10, 15, 20, 40, 5] \
Step3: [5, 10, 15, 20, 40] \
Worst Case= O(N^2)
Best Case=O(N)
Average Case=O(N^2)

```python
# Method 1
li=[20,10,40,15,5]
n=len(li)
count=0
for i in range(1,n):
    for j in range(0,i):
        if li[i]<li[j]:
            li[j],li[i]=li[i],li[j]
    print("Step"+str(count)+": " +str(li))
    count=count+1
    
print(li)
        
        
```

```python
# Method 2
li=[15,25,20,50,45,10,4]
print(li)
n=len(li)
for i in range(1,n):
    j=i-1
    temp=li[i]
    while (j>=0 and li[j]>temp):
        li[j+1]=li[j]

        j-=1
        
    li[j+1]=temp
    
    print(li)
        
print(li)   
```
