```python
def merge_algorithm(a, b):
    """Merge two sorted lists into a single sorted list."""
    left_idx = 0
    right_idx = 0
    k = 0
    c = [0] * (len(a) + len(b))

    # Merge elements from both lists
    while left_idx < len(a) and right_idx < len(b):
        if a[left_idx] <= b[right_idx]:
            c[k] = a[left_idx]
            left_idx += 1
        else:
            c[k] = b[right_idx]
            right_idx += 1
        k += 1

    # Copy remaining elements from a
    while left_idx < len(a):
        c[k] = a[left_idx]
        left_idx += 1
        k += 1

    # Copy remaining elements from b
    while right_idx < len(b):  # FIX: was using left_arr instead of right_idx
        c[k] = b[right_idx]
        right_idx += 1
        k += 1

    return c


def merge_sort(a):
    """Sort a list using the merge sort algorithm."""
    if len(a) <= 1:
        return a

    mid = len(a) // 2
    left = merge_sort(a[:mid])
    right = merge_sort(a[mid:])
    return merge_algorithm(left, right)


# Example usage
if __name__ == "__main__":
    arr = [38, 27, 43, 3, 9, 82, 10]
    sorted_arr = merge_sort(arr)
    print("Original array:", arr)
    print("Sorted array:", sorted_arr)
```
